```python
#Dannie Bryan March 10th 2023
#Burger Shop Submission

                           
        # Define a FoodItem class to hold menu items and order information

class FoodItem:
        # Define a dictionary of menu items with their prices for different categories
    menu = {               
        "burgers": {
            "kracken": 5.99,
            "double kracken": 7.99, 
            "chickracken": 6.99,
            "seaweed": 9.99,
            "none": 0.00
        },
        "sides": {
            "fries": 4.99,
            "krackenrings": 5.99,
            "salad": 6.99,
            "none": 0.00
        },
        "drinks": {
            "coke": 2.00,
            "sprite": 2.00,
            "rum": 5.00,
            "none": 0.00
        },
        "extras": {    
            "onions": 0.99,
            "pickles": 0.99,
            "mustard": 0.99,
            "ketchup": 0.99,
            "none": 0.00
        },
        "drink_sizes": {
            "medium": 0.99,
            "large": 1.99,
            "extra": 2.99,
            
        }
    }  
        # Initialize a total and list of items for each FoodItem object        
    total = 0
    items = []

        # Define a display method to print the order information
    def display(self):
        print("This here be the final treasure trove:")
        for item in self.items:
            for category in self.menu.keys():
                if item in self.menu[category].keys():
                    print(f"{category.capitalize()}: {item} (£{self.menu[category][item]:.2f})")
        print(f"Avast! This order is complete, ye owe me: £{self.order_total:.2f}, Spanish gold pieces only please.")


        #Welcome user and introduce system        
print("Welcome to Scurvey Burger, ye now stand amongst our ranks ")
print("Please use this here screen to place yar order. ")
print("We have booty full burgers, sides and drinks at low tide. ")
print("Let's walk the plank which be our ordering system! ")  
        
        # Define a Burger class as a subclass of FoodItem
class Burger(FoodItem):
    def __init__(self):
        # Call the FoodItem constructor        
        FoodItem.__init__(self)
        self

        # Define a method to order burgers    
    def order_burgers(self):
        while True:
        # Prompt the customer for their burger choice        
            item = input(f"Kindly choose yar burger Boatswain... {self.menu['burgers'].keys()}")
        # If the customer's choice is not valid, prompt them to try again        
            if item not in self.menu['burgers'].keys():
                print("Ahh ye be shipwrecked, try again")
                continue
        # Add the customer's choice to the order list and update the total cost        
            self.items.append(item)
            self.total += self.menu['burgers'][item]
        # Prompt the customer to order another burger or continue
            another = input("Will you be having another burger? Will ye be pressin' (y for yes)")
            if another =='y':
                continue
            else:                          
                break

        # Define an Extras (for the burger) class as sub class of FoodItem
class Extras(FoodItem):
    def __init__(self):
        # Call the FoodItem constructor
        FoodItem.__init__(self) 
        # Define a method to order Extras
    def order_extras(self):
        while True:
        # Prompt the customer for their (burger) extras choice  
            item = input(f"D'ya want some extra's thar? {self.menu['extras'].keys()}")
        # If the customer's choice is not valid, prompt them to try again   
            if item not in self.menu['extras'].keys():
                print("Sailed off the edge of the world ye did, try again")
                continue
            self.items.append(item)
            self.total += self.menu['extras'][item]
        # Prompt the customer to order another choice or continue
            another = input("Might ye be orderin' some more extras? Would ye be pressin' (y for yes)")
            if another =='y':
                continue
            else: 
                break
                
        # Define an Sides class as sub class of FoodItem                
class Sides(FoodItem):
    def __init__(self):
        # Call the FoodItem constructor
        FoodItem.__init__(self) 
        # Define a method to order sides
    def order_sides(self):
        while True:
        # Prompt the customer for their Sides choice  
            item = input(f"Arrr we reckon something here'll take your fancy... {self.menu['sides'].keys()}")
        # If the customer's choice is not valid, prompt them to try again   
            if item not in self.menu['sides'].keys():
                print("Cast ye net again, you caught nought...'")
                continue
            self.items.append(item)
            self.total += self.menu['sides'][item]
            # Prompt the customer to order another choice or continue
            another = input("Will you be needing another side? Kindly be pressin' (y for yes)")
            if another =='y':
                continue
            else: 
                break
                
                
        # Define an Drinks class as sub class of FoodItem                
class Drinks(FoodItem):
        # Call the FoodItem constructor
    def __init__(self):
        FoodItem.__init__(self) 
        # Define a method to order Drinks
    def order_drinks(self):
        while True:
        # Prompt the customer for their Drinks choice  
            item = input(f"Shiver me timbers, get some grog n'all... {self.menu['drinks'].keys()}")
         # If the customer's choice is not valid, prompt them to try again   
            if item not in self.menu['drinks'].keys():
                print("Avast, ye be sinking into the briney depths, have another try")
                continue
            self.items.append(item)
            self.total += self.menu['drinks'][item]
            # Prompt the customer to order another choice or continue
            another = input("Is it passible ye'll be wantin more to drink? Ye may be pressin' (y for yes)")
            if another =='y':
                continue
            else: 
                break
        # Define an Drink sizes class as sub class of FoodItem                
class Drink_sizes(FoodItem):
    def __init__(self):
        # Call the FoodItem constructor
        FoodItem.__init__(self) 
        # Define a method to order drink sizes
    def order_drink_size(self):
        while True:
        # Prompt the customer for their Drink sizes choice  
            item = input(f"How big a tankard shall we pull f'ye? {self.menu['drink_sizes'].keys()}")
        # If the customer's choice is not valid, prompt them to try again   
            if item not in self.menu['drink_sizes'].keys():
                print("Hornswaggled again, we'll give it another go!")
                continue
            self.items.append(item)
            self.total += self.menu['drink_sizes'][item]
            # Prompt the customer to order another choice or continue
            another = input("is there any more drinks to size? Ye'll be pressin' (y for yes)")
            if another =='y':
                continue
            else: 
                break
                
        # Define a new class called Order that inherits from the FoodItem class
class Order(FoodItem):
        # Define a constructor method for the Order class
    def __init__(self):
        # Call the constructor method of the parent class (FoodItem)
        # to initialize the food item attributes
        FoodItem.__init__(self)
        # Initialize the order_total attribute to 0
        self.order_total = 0

        # Define a method to take customer orders
    def take_order(self):
        # Create an instance of the Burger class and order burgers
        self.burgers = Burger()
        self.burgers.order_burgers()
        # Add the total cost of burgers to the order total
        self.order_total += self.burgers.total

        # Create an instance of the Extras class and order extras
        self.extras = Extras()
        self.extras.order_extras()
        # Add the total cost of extras to the order total
        self.order_total += self.extras.total

        # Create an instance of the Sides class and order sides
        self.sides = Sides()
        self.sides.order_sides()
        # Add the total cost of sides to the order total
        self.order_total += self.sides.total

        # Create an instance of the Drinks class and order drinks
        self.drinks = Drinks()
        self.drinks.order_drinks()
        # Add the total cost of drinks to the order total
        self.order_total += self.drinks.total

        # Create an instance of the Drink_sizes class and order drink sizes
        self.drink_sizes = Drink_sizes()
        self.drink_sizes.order_drink_size()
        # Add the total cost of drink sizes to the order total
        self.order_total += self.drink_sizes.total 

        # Create an instance of the Order class
order1 = Order() 
        # Take the order for the customer
order1.take_order()
        # Display the total cost of the order
order1.display()
```
